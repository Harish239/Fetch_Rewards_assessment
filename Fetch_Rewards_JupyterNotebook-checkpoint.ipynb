# Flattening the multiple object Json file and saving it as a excel file


```python
def json_excel(location, excel_file_name):

    entity_list = []
    for line in open(location, 'r'):
        entity_list.append(json.loads(line))
    
    entity_df = pd.DataFrame.from_dict(pd.json_normalize(entity_list), orient='columns')
    entity_df.to_excel("{}.xlsx".format(excel_file_name), header = True)
    print("Excel File created")
    return entity_df
```

```python
brands_df = json_excel('brands.json', 'brands')
users_df = json_excel('users.json', 'users')
```

# Flattening Multi level Json file(receipts.json)

```python
receipt_list = []
location = 'receipts.json'
for line in open(location, 'r'):
    receipt = flatten(json.loads(line))
    receipt_list.append(json.loads(line))
    


receipt_df = pd.DataFrame.from_dict(pd.json_normalize(receipt_list), orient='columns')

 
lst_unnest_dct = []
for row in range(len(receipt_df)):
    temp_id = receipt_df.loc[row, "_id.$oid"]
    
    lst_rewardsReceiptItemList = receipt_df.loc[row, "rewardsReceiptItemList"]
    try:
        for ele in lst_rewardsReceiptItemList:    
            ele["_id.$oid"] = temp_id
            lst_unnest_dct.append(ele)
    except:
        continue
    
unnested_rewardsReceiptItemList = pd.DataFrame(lst_unnest_dct)

unnest_receipt_df = receipt_df.merge(unnested_rewardsReceiptItemList, on = "_id.$oid", how = 'inner' )

unnest_receipt_df.to_excel("unnested_receipts.xlsx", header = True)
```

### Q3. Evaluate Data Quality Issues in the Data Provided

1) __rewardReceiptItemList__ doesnt have unique identifier to associate with the __brands__ table.<br>
2) Columns associated to __date__ are in __13 digit UNIX timestamp__ data type and it should be converted to standard __date__ data type.<br>
3) In order to join __rewardReceiptItemList__ with the __brands__ table, __brandCode__ is the joinable column and __brandCode__ is not __unique__ in the __brands__ table.<br>
4) Out of __1167__ rows in __brands__ table, there are only __898 unique values__ for __brandCode__ column <br>
5) Instead of using __brandCode__ as a Foreign key for __receipts__ table, it is better to use the ___id(brand table)__ to avoid duplicates.<br>
6) There is __no "Accepted"__ value in the __rewardReceiptStatus__ column.<br>
7) For __441 receipts__ there are __no rewardReceiptItemList__ present.<br>
8) __Users__ table has __212 unique__ users, while __receipts__ table __258__ users. __Foreign key constraint__ is violated.<br>
